```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
import numpy as np

def step(z):#defines a function named step that decides the output of the perceptron
    return 1 if z > 0 else 0  #If input z is greater than zero return 1 otherwise return 0

def perceptron(X, y):  #X is input data (features) & y is the actual label (target) 
    X = np.insert(X, 0, 1, axis=1)  # it will add 1 after the x 
    weights = np.ones(X.shape[1])  #  creates an array of weights .
    lr = 0.4             #learning rate 

    for i in range(1000):
        j = np.random.randint(0, len(X))   #Picks a random row index j from your data.
        y_hat = step(np.dot(X[j], weights))    #  for predicting the values either yes(1) or no(0)
        weights = weights + lr * (y[j] - y_hat) * X[j]   

    return weights[0], weights[1:]

# Example data
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 0, 1])

# Train perceptron
intercept_, coef_ = perceptron(X, y)  #Calling the perceptron
print("Coefficients:", coef_)
print("Intercept:", intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
