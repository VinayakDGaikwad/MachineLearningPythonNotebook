```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the input data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
```

```python
# Step 2: Construct the design matrix X with a bias (intercept) term
# Add a column of 1s for the intercept
X = np.column_stack((np.ones(len(x)), x))  # Shape: (n_samples, 2)
Y = y.reshape(-1, 1)                       # Make Y a column vector
```

```python
# Step 3: Apply the Normal Equation to compute theta = (X^T X)^-1 X^T y
XtX = X.T @ X
XtX_inv = np.linalg.inv(XtX)
XtY = X.T @ Y
theta = XtX_inv @ XtY  # theta = [b, w]

# Extract intercept (b) and slope (w)
b, w = theta[0, 0], theta[1, 0]

print(f"Fitted line: y = {w:.2f}x + {b:.2f}")

# Step 4: Predict values using the fitted line
y_pred = w * x + b
```

```python
# Step 5: Plot the result
plt.scatter(x, y, color='blue', label='Original data')
plt.plot(x, y_pred, color='red', label='Fitted line (Matrix Method)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression via Matrix Math')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Input data
x = np.array([1, 2, 3, 8])
y = np.array([2, 4, 5, 7,])

Y = y.reshape(-1, 1)

# Step 2: Settings
max_degree = 10  # you can increase this if needed
x_plot = np.linspace(min(x), max(x), 200)
mse_list = []

# Step 3: Create subplots for each polynomial fit
fig, axs = plt.subplots(2, (max_degree + 1) // 2, figsize=(18, 8))
axs = axs.flatten()  # flatten 2D array to 1D for easy indexing

for degree in range(1, max_degree + 1):
    # Step 4: Create design matrices
    X = np.vander(x, N=degree + 1, increasing=True)
    X_plot = np.vander(x_plot, N=degree + 1, increasing=True)

    # Step 5: Solve for theta using normal equation
    theta = np.linalg.pinv(X.T @ X) @ X.T @ Y

    # Step 6: Predictions
    y_pred = X @ theta
    y_plot = X_plot @ theta

    # Step 7: Calculate MSE
    mse = np.mean((y - y_pred.flatten())**2)
    mse_list.append(mse)

    # Step 8: Plot current fit
    ax = axs[degree - 1]
    ax.scatter(x, y, color='blue', label='Data')
    ax.plot(x_plot, y_plot, color='red', label=f'Degree {degree}')
    ax.set_title(f'Degree {degree} | MSE={mse:.2f}')
    ax.grid(True)
    ax.legend()

# Step 9: Add MSE vs Degree Plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, max_degree + 1), mse_list, marker='o', color='green')
plt.title("Error Analysis: MSE vs Polynomial Degree")
plt.xlabel("Polynomial Degree")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.show()
```
