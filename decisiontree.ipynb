```python
#Predict House Price Based on Size (sq.ft)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Step 1: Input data
X = np.array([[500], [1000], [1500], [2000], [2500]])     # House size
y = np.array([1500000, 2500000, 3500000, 4500000, 5500000])  # Price

# Step 2: Train model
model = DecisionTreeRegressor()
model.fit(X, y)

# Step 3: Predict price of 1800 sq.ft house
predicted_price = model.predict([[1800]])
print(f"Predicted price for 1800 sq.ft house is â‚¹{predicted_price[0]:,.2f}")

# Step 4: Plot
X_grid = np.arange(500, 2501, 10).reshape(-1, 1)
y_pred = model.predict(X_grid)

plt.scatter(X, y, color='green', label='Actual Prices')
plt.plot(X_grid, y_pred, color='blue', label='Predicted (Decision Tree)')
plt.title("House Price Prediction using Decision Tree Regression")
plt.xlabel("House Size (sq.ft)")
plt.ylabel("Price (â‚¹)")
plt.legend()
plt.grid(True)
plt.show()
```
